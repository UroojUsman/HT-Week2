```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
# Create a black image
img = cv2.imread('E://urooj desktop 2//Urooj.jpg',1)
thickness = 8
# Draw a diagonal blue line with thickness of 5 px
pts = np.array([[600,500],[400,100],[200,500]], np.int32)
pts = pts.reshape((-1,1,2))
image=cv2.polylines(img,[pts],True,(255,255,0),thickness)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))

```
